## Data Transformation

```python
# We'll convert the Date column to datetime format to be able to manipulate easier
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Extract the information to see the impact of the customer on a day
df['Month'] = df['Date'].dt.month # month that transaction was made
df['Year'] = df['Date'].dt.year # year that transaction was made
df['Weekday'] = df['Date'].dt.day_name() # The day of transaction
```

```python
# Removing any extra spaces and making text lower case to make it easier to compare in the future
df['ProductName'] = df['ProductName'].str.strip().str.lower()
df['ProductVendor'] = df['ProductVendor'].str.strip().str.lower()
```

```python
df.head(10)
```

### Purpose
- We cleaned up date column to change the format so we can extract specific information like the month, year and day of the week for future analysis 

## Understanding Customer Behavior 

```python
# Summarize customer behavior
customer_data = df.groupby('CustID').agg(
    TotalSpent=('OrderTotal', 'sum'),
    Transactions=('TransactionID', 'count'),
    AvgOrderCost=('OrderTotal', 'mean')
).sort_values(by='TotalSpent', ascending=False)
```

```python
customer_data.head(10)
```

```python
# Let's take a look at total spending by weekday
day_of_week_data = df.groupby('Weekday').agg(
    Total_Spent=('OrderTotal', 'sum'),
    Transactions=('TransactionID', 'count'),
    AvgOrderCost=('OrderTotal', 'mean')
).sort_values(by='Total_Spent',ascending=False)
```

```python
day_of_week_data
```

```python
# Plot to show 
day_of_week_data['Total_Spent'].plot(kind='bar', figsize=(10, 6), title='Total Revenue by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Revenue ($)')
plt.show()
```

```python
# Visualize the top 10 Cust by their total spending
customer_data.head(10).plot(kind='bar', y='TotalSpent', figsize=(10,6), title='Top 10 Customers')
plt.xlabel('Customer ID')
plt.ylabel('Total Spent ($)')
plt.xticks(rotation=0) # readability 
plt.show()
```

### Observations
- We have CustID 27 who made 37 transactions that had a total spent of 16,291; Showing the extent to how much revenue they generate
- We can also see that the high spending customers tend to value premium quality as their average order values are above the median cost of 350
- We also see that Wed has the highest revenue per the weekday and Sat is the lowest

## Product Highlights

```python
# Group products by ProductName
product_data_grouped = df.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False).head(10)
```

```python
# USe histogram of the top products from product_data
fig, ax = plt.subplots(figsize=(10, 6)) 
# horizonatal bar for better readability
ax.barh(   
    product_data_grouped.head(10).index,
    product_data_grouped.head(10).values,
    color='skyblue'
)
ax.set_title("Top 10 Products by Quantity Ordered")
ax.set_xlabel("Quantity Ordered")
ax.set_ylabel("Product Name")
plt.show()
```

### Observations
- Gnocchi is the highest ordered item, and clam chowder and guaran.
- We can focus on increasing quantity or more promotion to increase overall revenue
- On the other hand, we can focus on the lower quantity ordered of the top 10 list and learn how to promote it.
