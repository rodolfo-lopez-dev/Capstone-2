## Data Transformation

```python
# We'll convert the Date column to datetime format to be able to manipulate time-based analysis 
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Extract the information to see the impact of the customer on a day
df['Month'] = df['Date'].dt.month # month that transaction was made
df['Year'] = df['Date'].dt.year # year that transaction was made
df['Weekday'] = df['Date'].dt.day_name() # The day of transaction
```

```python
# Removing any extra spaces and making text lower case to make it easier to compare in the future
df['ProductName'] = df['ProductName'].str.strip().str.lower()
df['ProductVendor'] = df['ProductVendor'].str.strip().str.lower()
```

```python
df.head(10)
```

### Purpose
- We cleaned up date column to change the format so we can extract specific information like the month, year and day of the week for future analysis 

## Understanding Customer Behavior 

```python
# Summarize customer behavior
customer_data = df.groupby('CustID').agg(
    TotalSpent=('OrderTotal', 'sum'),
    Transactions=('TransactionID', 'count'),
    AvgOrderCost=('OrderTotal', 'mean')
).sort_values(by='TotalSpent', ascending=False)
```
