```python
df1 = pd.read_csv('customer_list_updated.csv', delimiter='|') # specify delimiter ('|') to separate the single column to it's own
df1.columns =['Cust_id', 'Date', 'Time', 'Name', 'Email', 'Phone', 'Sms_opt_out']
df1
```

## Clean Data

### Adjusting the name column

```python
# cleaning "name" column from unwanted characters
df1['Name'] = df1['Name'].str.replace(r"[^A-Za-z\-\.\s]", "", regex=True)
df1
```

### Showing all rows

```python
pd.options.display.max_rows = None
df1
```

### Adjusting the phone number formatting 

```python
# focusing on the phone number that includes '1'
df1[284:296]
```

```python
# focus on setting the phone number only in NNN-NNN-NNNN format
df1['Phone'] = (
    df1['Phone']
    .astype(str)  # Convert to string
    .str.replace(r'\D', '', regex=True)  # Remove non-numeric characters
    .str[-10:]  # only using the last 10 digits (to handle leading 1 or excess digits)
    .str.replace(r'(\d{3})(\d{3})(\d{4})', r'\1-\2-\3', regex=True)  #ensuring format is NNN-NNN-NNNN
)
# this should fix the extra digit issue
df1
```

```python
# now addressing the row 300
df1['Phone'] = df1['Phone'].replace('', "Missing Number")
df1[300:301:1]
```

### SMS  

```python
# From a marketing perspective, we want to maximize our reach and potential customer base. By including all participants and non-participants, we can expand valuable opportunities for business growth.
df1['Sms_opt_out'] = df1['Sms_opt_out'].fillna('N')
df1
```

```python
# This shows how many Na values are left
# df1['Sms_opt_out'].isna().any(axis=1) where to find Na values and where
df1['Sms_opt_out'].isna().sum()
```
